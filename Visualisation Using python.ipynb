```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
pokemon=pd.read_csv('C:/Users/UditKhandelwal/Desktop/upGrad/Datasets/pokemon.csv')
pokemon.head(10)
```

```python
#df=pokemon.HP
#df
```

```python
pokemon_new=pokemon[['Index','Name','Type1','HP','Attack','Defense','SpAtk','SpDef','Speed']]
#df=pokemon.pivot_table(index=['Type1','Name'],values=['HP','Attack','Defense','SpAtk','SpDef','Speed']
pokemon_new1=pokemon_new.groupby(['Type1'],as_index=False)[['HP','Attack']].mean()
pokemon_new1.head(10)
```

```python
plt.figure(figsize=(20,8))
plt.plot(pokemon_new1['Type1'],pokemon_new1[['HP','Attack']])
plt.xlabel('Type')
plt.ylabel('Strength')
plt.legend(['HP','Attack'])

plt.show()
```

```python
plt.scatter(pokemon['HP'],pokemon['Attack'],c="r",alpha=0.5,edgecolors="g")
plt.xlabel('HP')
plt.ylabel('Attack')
plt.show()
```

```python
sns.lmplot('HP','Attack',data=pokemon,fit_reg=False,row='Type1',hue='Type1' )
#plt.plot(pokemon['HP'].mean(),pokemon['Attack'].mean(),c="g")
sns.lmplot('HP','Attack',data=pokemon,fit_reg=False,hue='Type2')
```
